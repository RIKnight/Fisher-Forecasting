```python
# put the fun in functions

def getSigmas(Fij):
    #print np.diag(np.linalg.inv(Fij))
    return np.sqrt(np.diag(np.linalg.inv(Fij)))
def fixParam(Fij,pNum):
    return np.delete(np.delete(Fij,pNum,0),pNum,1)

# define a few functions for calculating FoM

# define function for getting ap (pivot a) for each matrix
def getAp(FijInv,iw0=8,iwa=7):
    dw0dwa = FijInv[iw0,iwa]
    dwadwa = FijInv[iwa,iwa]
    return 1+dw0dwa/dwadwa

# define function for getting sigma^2(w_p) (variance of w_p)
def getVarWp(FijInv,iw0=8,iwa=7):
    dw0dw0 = FijInv[iw0,iw0]
    dw0dwa = FijInv[iw0,iwa]
    dwadwa = FijInv[iwa,iwa]
    a_p = getAp(FijInv,iw0,iwa)
    return dw0dw0 +2*(1-a_p)*dw0dwa +(1-a_p)**2*dwadwa

# define function for getting value proportional to FoM
def FoMprop(FijInv,iw0=8,iwa=7):
    sigmaWp = np.sqrt(getVarWp(FijInv,iw0,iwa))
    sigmaWa = np.sqrt(FijInv[iwa,iwa])
    return 1/(sigmaWp*sigmaWa)
```

```python
def getFoMs(FijBins,fixNum=6,w0Num=6,waNum=7):
    """
    FijBins: an array of Fisher matrices
    fixNum: the paramNum to fix
    w0Num: the index of w0 after fixing
    waNum: the index of wa after fixing
    returns:
    
    """
    nFij = FijBins.__len__()
    myFoMs = np.zeros(nFij)
    for FijNum in range(nFij):
        myFij = fixParam(FijBins[FijNum],fixNum)
        invFij = np.linalg.inv(myFij)
        myFoMs[FijNum] = FoMprop(invFij,iw0=w0Num,iwa=waNum)
    return myFoMs
```

```python
# the arrays of Fisher matrices for making FoM plots:
# (2 for Gold, 4 for Optimistic)
"""
FijBins2_01_lin
FijBins2_01g_lin
FijBins2_02_lin
FijBins2_02g_lin

FijBins4_01_lin
FijBins4_01g_lin
FijBins4_02_lin
FijBins4_02g_lin

FijBins2_01_nln
FijBins2_01g_nln
FijBins2_02_nln
FijBins2_02g_nln

FijBins4_01_nln
FijBins4_01g_nln
FijBins4_02_nln
FijBins4_02g_nln
"""

# define function for adding prior
def addPrior(Fij,prior,tauIndex=5):
    """
    Inputs:
        Fij: the Fisher matrix
        prior: the 1-sigma constraint to add
        tauIndex: the index of the parameter to add constraint to
    Returns:
        The modified Fisher matrix; same dimensions, ordering, etc. as Fij
    """
    toAdd = prior**(-2)
    myFij = np.copy(Fij)
    myFij[tauIndex,tauIndex] += toAdd
    return myFij
    
# how about one for adding prior to a list of matrices?
def addPriorToList(FijList,prior,tauIndex=5):
    """
    FijList: a list of Fisher matrices.  Must all have same parameter at tauIndex.
    """
    #nFijs = FijList.__len__()
    returnList = []
    for Fij in FijList:
        returnList.append(addPrior(Fij,prior,tauIndex=tauIndex))
    return returnList
```

```python
# define a few functions for calculating FoM

# define function for getting ap (pivot a) for each matrix
def getAp(FijInv,iw0=8,iwa=7):
    dw0dwa = FijInv[iw0,iwa]
    dwadwa = FijInv[iwa,iwa]
    return 1+dw0dwa/dwadwa

# define function for getting sigma^2(w_p) (variance of w_p)
def getVarWp(FijInv,iw0=8,iwa=7):
    dw0dw0 = FijInv[iw0,iw0]
    dw0dwa = FijInv[iw0,iwa]
    dwadwa = FijInv[iwa,iwa]
    a_p = getAp(FijInv,iw0,iwa)
    return dw0dw0 +2*(1-a_p)*dw0dwa +(1-a_p)**2*dwadwa

# define function for getting value proportional to FoM
def FoMprop(FijInv,iw0=8,iwa=7):
    sigmaWp = np.sqrt(getVarWp(FijInv,iw0,iwa))
    sigmaWa = np.sqrt(FijInv[iwa,iwa])
    return 1/(sigmaWp*sigmaWa)
```

```python
# get FoMs
def getFoMPropBins(FijList):
    returnList = []
    for Fij in FijList:
        returnList.append(FoMprop(np.linalg.inv(Fij)))
    return np.array(returnList)

FoMPropBins2_01_lin_tA = getFoMPropBins(FijBins2_01_lin_tA)
FoMPropBins2_01_lin_tB = getFoMPropBins(FijBins2_01_lin_tB)
FoMPropBins2_01_lin_tC = getFoMPropBins(FijBins2_01_lin_tC)

FoMPropBins2_02_lin_tA = getFoMPropBins(FijBins2_02_lin_tA)
FoMPropBins2_02_lin_tB = getFoMPropBins(FijBins2_02_lin_tB)
FoMPropBins2_02_lin_tC = getFoMPropBins(FijBins2_02_lin_tC)

FoMPropBins4_01_lin_tA = getFoMPropBins(FijBins4_01_lin_tA)
FoMPropBins4_01_lin_tB = getFoMPropBins(FijBins4_01_lin_tB)
FoMPropBins4_01_lin_tC = getFoMPropBins(FijBins4_01_lin_tC)

FoMPropBins4_02_lin_tA = getFoMPropBins(FijBins4_02_lin_tA)
FoMPropBins4_02_lin_tB = getFoMPropBins(FijBins4_02_lin_tB)
FoMPropBins4_02_lin_tC = getFoMPropBins(FijBins4_02_lin_tC)


FoMPropBins2_01_nln_tA = getFoMPropBins(FijBins2_01_nln_tA)
FoMPropBins2_01_nln_tB = getFoMPropBins(FijBins2_01_nln_tB)
FoMPropBins2_01_nln_tC = getFoMPropBins(FijBins2_01_nln_tC)

FoMPropBins2_02_nln_tA = getFoMPropBins(FijBins2_02_nln_tA)
FoMPropBins2_02_nln_tB = getFoMPropBins(FijBins2_02_nln_tB)
FoMPropBins2_02_nln_tC = getFoMPropBins(FijBins2_02_nln_tC)

FoMPropBins4_01_nln_tA = getFoMPropBins(FijBins4_01_nln_tA)
FoMPropBins4_01_nln_tB = getFoMPropBins(FijBins4_01_nln_tB)
FoMPropBins4_01_nln_tC = getFoMPropBins(FijBins4_01_nln_tC)

FoMPropBins4_02_nln_tA = getFoMPropBins(FijBins4_02_nln_tA)
FoMPropBins4_02_nln_tB = getFoMPropBins(FijBins4_02_nln_tB)
FoMPropBins4_02_nln_tC = getFoMPropBins(FijBins4_02_nln_tC)

print 'done.'
```
