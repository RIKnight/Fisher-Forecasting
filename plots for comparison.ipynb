# define a function that will calculate the full F_ijl
Caluclate Fisher matrix for special simplified case of:
                nonoverlapping redshift bins
                dN/dz = WinK

## the following function assumes that the covariance matrix of the observables is diagonal
## however, it actually is not.
## Q: how does this change the analysis?

```python
# this needs three things, probably from Fobj
# what are the properties of the current Fobj?

def getFijl(biases,crossCls,ells):
    """
        Purpose:
            Caluclate Fisher matrix for special simplified case of:
                nonoverlapping redshift bins
                dN/dz = WinK
        Note: to calculate Full Fij, sum Fijl over l
        Inputs:
            biases: the fiducial biases used in the calculation for each bin
            crossCls: all cross power spectra, implemented as in FisherMatrix.crossCls
                has shape (nMaps,nMaps,lmax-1)  where -1 is to omit ell=1
                Kappa map is in 0 position, galaxy maps are in order after that
            ells: the ell values to include in array
        Returns:
            a 3 dimensional array (nBins,nBins,nElls)
    """
    nBins = biases.__len__()
    nElls = ells.__len__()
    print 'nBins, nElls: ',nBins,nElls
    print 'ells: ',ells
    
    # create array of zeros
    Fijl = np.zeros((nBins,nBins,nElls))
    
    # pull out some needed cross spectra
    Ckk = crossCls[0,0,:]
    
    # fill up array via loops
    for i in range(nBins):
        Cgigi = crossCls[i+1,i+1,:] # +1 since crossCls includes kappa, but Fij does not
        Ckgi  = crossCls[0  ,i+1,:]
        for j in range(nBins):
            Cgjgj = crossCls[j+1,j+1,:]
            Cgigj = crossCls[i+1,j+1,:]
            Ckgj  = crossCls[0  ,j+1,:]
            for ellI in range(nElls):
                ell = ellI+2 # since ell=0,1 not included
                # the gigi gjgj part
                if i == j:
                    Fijl[i,j,ellI] += 2./biases[i]**2 *(2*ell+1)
                # the kgi kgj part
                Fijl[i,j,ellI] += Cgigi[ellI]/biases[i]**2 * Cgjgj[ellI]/biases[j]**2 * \
                    (2*ell+1)/(Ckk[ellI]*Cgigj[ellI] + Ckgj[ellI]*Ckgi[ellI])
                # the kgi gjgj part
                if i == j:
                    Fijl[i,j,ellI] += 2.*(2*ell+1)/biases[i]**3 * Cgigi[ellI]/Ckgi[ellI]
                # the kk kgj, kk gjgj, and kk kk parts are all zero, since dCl^kk/db = 0 
    
    return Fijl

Fijl = getFijl(Fobj.binBs,Fobj.crossCls,Fobj.ells)
```
